```python
class Student:
  def __init__(self, student_ID, name, program, sex, weight, height):
    self.student_ID = student_ID
    self.name = name
    self.program = program
    self.sex = sex
    self.weight = weight
    self.height = height
    self.courses = {}

  def enroll_course(self, course_name):
    self.courses[course_name] = {'hands_on': 0, 'quiz': 0, 'seatwork': 0, 'assignment': 0, 'recitation' : 0, 'midterm': 0, 'final': 0, 'class_standing': 0, 'final_grade': None}

  def upgrade_grade(self, course_name, hands_on, quiz, seatwork, assignment, recitation, midterm, final):
    if course_name in self.courses:
        class_standing = calculate_class_standing(hands_on, quiz, assignment)
        final_grade = calculate_final_grade(final, class_standing, self.courses.get(course_name, {}).get('final_grade'))

    self.courses[course_name] = {
    'hands_on': hands_on,
    'quiz': quiz,
    'seatwork' : seatwork,
    'assignment' : assignment,
    'recitation' : recitation,
    'midterm' : midterm,
    'final' : final,
    'class_standing' : class_standing,
    'final_grade' : final_grade
    }

  def compute_final_grade(self, course_name):
    if course_name in self.courses:
        return self.courses[course_name]['final_grade']
    return None

  def view_courses(self):
    for course, grades in self.courses.items():
      print(f"course: {course}, final grade: {self.compute_final_grade(course):.2f}")

courses_list = ["CPE 103", "CPE 103L", "GEC 002", "GEC 002", "GEC 003", "IE 102",
                "MAT 002", "NSTP 122", "PATHFIT 2", "PHY 001L", "PHY 001"
                ]


def calculate_class_standing(hands_on, quiz, assignment):
  return (hands_on * 0.5) + (quiz * 0.3) + (assignment * 0.2)

def calculate_final_grade(exam_grade, class_standing, prev_grade =  None):
  if prev_grade is None:
    return (exam_grade * 0.5) + (class_standing * 0.5)
  return (prev_grade * 1/3) + ((exam_grade * 0.5 + class_standing * 0.5) * 2/3)

def convert_to_ucc_grade(final_grade):
  if final_grade >= 98:
        return 1.00
  elif final_grade >= 95:
       return 1.25
  elif final_grade >= 90:
        return 1.50
  elif final_grade >= 85:
        return 1.75
  elif final_grade >= 80:
        return 2.00
  elif final_grade >= 75:
        return 2.25
  elif final_grade >= 70:
        return 2.50
  elif final_grade >= 65:
        return 2.75
  elif final_grade >= 60:
        return 3.00
  else:
        return 5.00

def get_student_info():
  student_ID = input("Enter the student ID: ")
  name = input("Enter the student name: ")
  program = input("Enter the student program: ")
  sex = input("Enter the student sex (M/F): ")
  weight = float(input("Enter the student weight (kg): "))
  height = float(input("Enter the student height (cm): "))
  return  student_ID, name, program, sex, weight, height

students = []
for _ in range(3):
  student_info = get_student_info()
  student = Student(*student_info)
  students.append(student)

def enroll_and_upgrade(student):
  print("Available courses:")
  for i, course in enumerate(courses_list, start = 1):
    print(f"{i}. {course}")
  course_choice = int(input(f"Enter a selected course number for {student.name}: ")) -1
  if 0 <= course_choice < len(courses_list):
      course_name = courses_list[course_choice]
      student.enroll_course(course_name)
      hands_on = float(input("Enter hands_on grade: "))
      quiz = float(input("Enter quiz grade: "))
      seatwork = float(input("Enter seatwork grade: "))
      assignment = float(input("Enter assignment grade: "))
      recitation = float(input("Enter recitation grade: "))
      midterm = float(input("Enter midterm grade: "))
      final = float(input("Enter final grade: "))
      student.upgrade_grade(course_name, hands_on, quiz, seatwork, assignment, recitation, midterm, final) # Fixed function call to upgrade_grade
      student.view_courses()
      # Calculate and print UCC grade within the function where course_name is defined
      ucc_grade = convert_to_ucc_grade(student.compute_final_grade(course_name))
      print(f"UCC Grade: {ucc_grade:.2f}")
  else:
    print("Invalid course selection.")

# Removed the lines causing the error as they were outside the function's scope
#ucc_grade = convert_to_ucc_grade(student.compute_final_grade(course_name))
#print(f"UCC Grade: {ucc_grade:.2f}")

for student in students:
    enroll_and_upgrade(student)
```
