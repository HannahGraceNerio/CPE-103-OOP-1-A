```python
# Function to calculate class standing
def calculate_class_standing(hands_on, quiz, assignment):
    return (hands_on * 0.5) + (quiz * 0.3) + (assignment * 0.2)

# Function to calculate grade at each
def calculate_grade(exam, class_standing, prev_grade=None):
    if prev_grade is None:
        return (exam * 0.5) + (class_standing * 0.5)
    return (prev_grade * 1 / 3) + ((exam * 0.5 + class_standing * 0.5) * 2 / 3)

# Function to convert numerical grade to UCC grading system
def convert_to_ucc_grade(final_grade):
    if final_grade >= 99:
        return 1.00
    elif final_grade >= 95:
        return 1.25
    elif final_grade >= 90:
        return 1.50
    elif final_grade >= 85:
        return 1.75
    elif final_grade >= 80:
        return 2.00
    elif final_grade >= 75:
        return 2.25
    elif final_grade >= 70:
        return 2.50
    elif final_grade >= 65:
        return 2.75
    elif final_grade >= 60:
        return 3.00
    else:
        return 5.00

# Get input for multiple students
students = []
num_students = 3  # Test for 3 students

for i in range(num_students):
    print(f"\nEnter details for Student {i + 1}")
    name = (input("Enter Student Name: "))

    # Prelim
    prelim_exam = float(input("Prelim Exam: "))
    prelim_ho = float(input("Prelim Hands-on Activity: "))
    prelim_quiz = float(input("Prelim Quiz: "))
    prelim_assignment = float(input("Prelim Assignment: "))

    prelim_cs = calculate_class_standing(prelim_ho, prelim_quiz, prelim_assignment)
    prelim_grade = calculate_grade(prelim_exam, prelim_cs)

    # Midterm
    midterm_exam = float(input("Midterm Exam: "))
    midterm_ho = float(input("Midterm Hands-on Activity: "))
    midterm_quiz = float(input("Midterm Quiz: "))
    midterm_assignment = float(input("Midterm Assignment: "))

    midterm_cs = calculate_class_standing(midterm_ho, midterm_quiz, midterm_assignment)
    midterm_grade = calculate_grade(midterm_exam, midterm_cs, prelim_grade)

    # Final
    final_exam = float(input("Final Exam: "))
    final_ho = float(input("Final Hands-on Activity: "))
    final_quiz = float(input("Final Quiz: "))
    final_assignment = float(input("Final Assignment: "))

    final_cs = calculate_class_standing(final_ho, final_quiz, final_assignment)
    final_grade = calculate_grade(final_exam, final_cs, midterm_grade)


    # Convert to UCC grade
    ucc_grade = convert_to_ucc_grade(final_grade)

    # Store student data
    students.append((name, prelim_grade, midterm_grade, final_grade, ucc_grade))

# Display results
print("\n--- Student Grade Summary ---")
for student in students:
    name, prelim, midterm, final, ucc = student
    print(f"\nStudent Name: {name}")
    print(f"Prelim Grade: {prelim:.2f}")
    print(f"Midterm Grade: {midterm:.2f}")
    print(f"Final Grade: {final:.2f}")
    print(f"UCC Grade: {ucc}")
```
