```python
class Character():
  def __init__(self, username):
    self.__username = username
    self.__hp = 100
    self.__mana = 100
    self.__damage = 5
    self.__str = 0
    self.__vit = 0
    self.__int = 0
    self.__agi = 0
  def getUsername(self):
    return self.__username
  def setUsername(self, new_username):
    self.__username = new_username
  def getHp(self):
    return self.__hp
  def setHp(self, new_hp):
    self.__hp = new_hp
  def getDamage(self):
    return self.__damage
  def setDamage(self, new_damage):
    self.__damage = new_damage
  def getStr(self):
    return self.__str
  def setStr(self, new_str):
    self.__str = new_str
  def getVit(self):
    return self.__vit
  def setVit(self, new_vit):
    self.__vit = new_vit
  def getInt(self):
    return self.__int
  def setInt(self, new_int):
    self.__int = new_int
  def getAgi(self):
    return self.__agi
  def setAgi(self, new_agi):
    self.__agi = new_agi
  def reduceHp (self, damage_amount):
    self.__hp = self.__hp - damage_amount
  def addHp(self, heal_amount):
    self.__hp = self.__hp + heal_amount

character1 = Character("Hannah")

print(character1.getUsername())
```

```python
class Novice(Character):
    def basicAttack(self, character):
        character.reduceHp(self.getDamage())
        print(f"{self.getUsername()} performed Basic Attack! Damage: {self.getDamage()}")


character1 = Novice("Novice")


print(character1.getUsername())
print(character1.getHp())
```

```python
class Swordsman (Novice):
  def __init__(self, username):
    super().__init__(username)
    self.setStr(5)
    self.setVit(10)
    self.setHp(self.getHp()+self.getVit())


  def slashAttack(self, character):
    self.new_damage = self.getDamage()+self.getStr()
    character.reduceHp(self.new_damage)
    print(f"{self.getUsername()} performed Slash Attack! -{self.new_damage}")
```

```python
import random
class Archer (Novice):
  def __init__(self, username):
    super().__init__(username)
    self.setAgi(5)
    self.setInt(5)
    self.setVit(5)
    self.setHp(self.getHp()+self.getVit())
  def rangedAttack(self, character):
    self.new_damage = self.getDamage()+random.randint(0,self.getInt())
    character.reduceHp(self.new_damage)
    print(f"{self.getUsername()} performed Slash Attack! -{self.new_damage}")
```

```python
class Magician (Novice):
  def __init__(self, username):
    super().__init__(username)
    self.setInt(10)
    self.setVit(5)
    self.setHp(self.getHp()+self.getVit())
  def heal(self):
    self.addHp(self.getInt())
    print(f"{self.getUsername()} performed Heal! +{self.getInt()}")
  def magicAttack(self, character):
    self.new_damage = self.getDamage()+self.getInt()
    character.reduceHp(self.new_damage)
    print(f"{self.getUsername()} performed Magic Attack! -{self.new_damage}")
```

```python
import random

# Character base class
class Character:
    def __init__(self, username):
        self.__username = username
        self.__hp = 100

    def getUsername(self):
        return self.__username

    def getHp(self):
        return self.__hp

    def setHp(self, hp):
        self.__hp = max(0, hp)

    def is_alive(self):
        return self.__hp > 0

# Role Classes
class Novice(Character):
    def attack(self, opponent):
        damage = random.randint(10, 20)
        opponent.setHp(opponent.getHp() - damage)
        print(f"{self.getUsername()} attacks {opponent.getUsername()} for {damage} damage.")

class Swordsman(Novice):
    def attack(self, opponent):
        damage = random.randint(15, 25)
        opponent.setHp(opponent.getHp() - damage)
        print(f"{self.getUsername()} uses Slash on {opponent.getUsername()} for {damage} damage.")

class Archer(Novice):
    def attack(self, opponent):
        damage = random.randint(12, 22)
        opponent.setHp(opponent.getHp() - damage)
        print(f"{self.getUsername()} uses Arrow Shot on {opponent.getUsername()} for {damage} damage.")

class Magician(Novice):
    def attack(self, opponent):
        damage = random.randint(18, 28)
        opponent.setHp(opponent.getHp() - damage)
        print(f"{self.getUsername()} casts Fireball on {opponent.getUsername()} for {damage} damage.")

class Boss(Swordsman, Archer, Magician):
    def __init__(self, username):
        super().__init__(username)

    def attack(self, opponent):
        damage = random.randint(10, 30)
        opponent.setHp(opponent.getHp() - damage)
        print(f"{self.getUsername()} (Boss) attacks {opponent.getUsername()} for {damage} damage.")

# Role selector
def choose_class(name, role="Novice"):
    if role == "Swordsman":
        return Swordsman(name)
    elif role == "Archer":
        return Archer(name)
    elif role == "Magician":
        return Magician(name)
    else:
        return Novice(name)

# Battle function
def battle(player1, player2):
    print(f"\n--- BATTLE START: {player1.getUsername()} vs {player2.getUsername()} ---")
    while player1.is_alive() and player2.is_alive():
        attacker = random.choice([player1, player2])
        defender = player2 if attacker == player1 else player1
        attacker.attack(defender)
        print(f"{player1.getUsername()} HP: {player1.getHp()} | {player2.getUsername()} HP: {player2.getHp()}")
        print("-" * 40)

    winner = player1 if player1.is_alive() else player2
    print(f"{winner.getUsername()} wins the match!")
    return winner.getUsername()

# Single Player Mode
def run_single_player():
    wins = 0
    for match in range(3):
        role = "Novice" if wins < 2 else random.choice(["Swordsman", "Archer", "Magician"])
        print(f"\nMatch {match + 1} | Player Role: {role}")
        player = choose_class("Player1", role)
        boss = Boss("Monster")
        winner = battle(player, boss)
        if winner == "Player1":
            wins += 1
    print(f"\nPlayer1 Total Wins: {wins}")

# Player vs Player Mode
def run_pvp():
    p1_wins = 0
    p2_wins = 0
    roles = ["Novice", "Swordsman", "Archer", "Magician"]

    for match in range(3):
        role1 = random.choice(roles)
        role2 = random.choice(roles)
        print(f"\nMatch {match + 1}")
        print(f"Player1 Role: {role1}")
        print(f"Player2 Role: {role2}")
        player1 = choose_class("Player1", role1)
        player2 = choose_class("Player2", role2)
        winner = battle(player1, player2)
        if winner == "Player1":
            p1_wins += 1
        else:
            p2_wins += 1

    print(f"\nPlayer1 Wins: {p1_wins} | Player2 Wins: {p2_wins}")

# Main Game Launcher
def run_full_demo():
    print("==== SINGLE PLAYER MODE ====")
    run_single_player()
    print("\n==== PLAYER VS PLAYER MODE ====")
    run_pvp()

# Run the complete demo
run_full_demo()
```
